【問題3】小麦の数の合計
8×8マスのチェス盤の小麦の数の合計を求めてください。また、各列の平均を求め、それを棒グラフで可視化してください。



```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel("column")
plt.ylabel("number")
plt.title("number in each column")
plt.bar(np.arange(1,4), np.array([1, 2, 3]))
plt.show()
```

```python
def ChessNum(n,m):
  n_squares2 = n * m
  small_board_list2 = [1]
  for _ in range(n_squares2 - 1):
    small_board_list2.append(2*small_board_list2[-1])
    import numpy as np
    small_board_ndarray2 = np.array(small_board_list2)  
  return [sum(small_board_ndarray2) ,small_board_ndarray2.reshape(n, m)]

X = ChessNum(8,8)[0]
Y = ChessNum(8,8)[1]

print("64マスの板に小麦を並べる合計は、：{}個である" .format(X))
#列ごとの平均は、
print(np.average(Y,axis=0))

import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel("column")
plt.ylabel("number")
plt.title("number in each column")
plt.bar(np.arange(1,9), np.average(Y,axis=0))
plt.show()
```

【問題5】後半は前半の何倍か
チェス盤の後半にある小麦の数は前半にある小麦の数の何倍かを求めてください。前半とは配列の0から3行目、後半は4から7行目です。

```python
def ChessNum(n1,n2):
  n_squares2 = n1 * n2
  small_board_list2 = [1]
  for _ in range(n_squares2 - 1):
    small_board_list2.append(2*small_board_list2[-1])
  import numpy as np
  small_board_ndarray2 = np.array(small_board_list2)
  a = small_board_ndarray2.reshape(8, 8)
  #前半の小麦の数　a1,後半のの小麦の数　a2,後半の数(a2)/前半の数(a1)＝Cと置く
  a1 = a[0:4,0:9].sum()
  a2 = a[4:9,0:9].sum()
  C = a2 / a1 
  return C

import math
C = ChessNum(8,8)
print("後半にある小麦の数は前半にある小麦の数の、：{}倍である" .format(math.floor(C)))

```

numpyのブロードキャスト》
概要：
pythonのnumpyには「ブロードキャスト」という仕様があります。
これはnumpyのベクトル・行列計算において、２入力間で次元数やサイズの小さい方を複製してデータを自動で統一する機能です。ブロードキャストを使うメリットとして、①コードが短くなるため、実行速度が早くなる。②メモリの使用量を節約できる。の２点が主に挙げられます。

```python
import numpy as np
a = np.array([0,1,2])
#ブロードキャストを使わない場合
b = np.array([5,5,5])
print(a + b)  # Out: [5,6,7]

#ブロードキャストを使う場合
print(a + 5)  # Out: [5,6,7]　※5が自動的に(1,3)の行列([5,5,5])に変換されている
```

```python
import numpy as np
def ChessNum2(n,m):
  n_squares = n * m
  small_board_ndarray = np.array([1]).astype(np.uint64)
  for _ in range(n_squares - 1):
      small_board_ndarray = np.append(small_board_ndarray, 2*small_board_ndarray[-1])
  print("64マスの板に小麦を並べる（ndarray）：{}".format(small_board_ndarray))

ChessNum2(8,8)
```
