```python
# #特徴量すべてを組み合わせを一度に表示させた、散布図行列を作る。
import matplotlib.pyplot as plt
import numpy as np
import seaborn
iris = seaborn.load_dataset( 'iris' )
# print(iris)
# seaborn.jointplot('sepal_length', 'sepal_width', data=iris)
# x=iris['sepal_length'], y=iris['sepal_width']
#,z=iris['petal_length'], z2=iris['petal_width'],z3=iris['species']
# c = np.c_[x,y]
# print(c)

import numpy as np

import matplotlib.pyplot as plt

z3=iris['species']
y1=iris['sepal_length']
y2=iris['sepal_width']
y3=iris['petal_length']
y4=iris['petal_width']
plt.scatter(z3, y1, label='y1', color='aqua')
plt.scatter(z3, y2, label='y2', color='olive')
plt.scatter(z3, y3, label='y3', color='red')
plt.scatter(z3, y4, label='y4', color='black')


plt.legend()

plt.show()
```

```python
#４つの特徴量に関する相関係数行列を作る
#相関係数行列はDataFrameのcorr()メソッドを使うことで求められます。
import pandas as pd
import scipy.stats as sts

heacet = load_iris()
heacet_data = pd.DataFrame(heacet.data, columns=["sepal_length","sepal_width","petal_length","petal_width"])
heacet_all = pd.concat([heacet_data], axis=1)
corr_mat = np.zeros((heacet_all.shape[1], heacet_all.shape[1]))

for i in range(heacet_all.shape[1]):    
    for j in range(heacet_all.shape[1]):        
        corrtest = sts.pearsonr(heacet_all[heacet_all.columns[i]], heacet_all[heacet_all.columns[j]])
        # stats配下にspearmanrやkendalltauも存在
        corr_mat[i,j] = corrtest[0]

df_corr = pd.DataFrame(corr_mat, columns=heacet_all.columns, index=heacet_all.columns)
print(df_corr)
```

```python
#相関係数行列のヒートマップを作る(１つ)
import seaborn as sns
from sklearn import datasets
dataset = datasets.load_iris()

features = dataset.data
feature_names = dataset.feature_names

    # N 行 M 列を M 行 N 列に変換して相関行列を計算する
correlation_matrix = np.corrcoef(features.transpose())

    # 相関行列のヒートマップを描く
sns.heatmap(correlation_matrix, annot=True,
                xticklabels=feature_names,
                yticklabels=feature_names)

    # グラフを表示する
plt.show()

```
