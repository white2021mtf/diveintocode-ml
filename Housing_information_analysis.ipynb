【問題6】分布の確認
seabornのsns.displot()やsns.histplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
目的変数に対して対数変換を行う。
対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。


seaborn.displot — seaborn 0.11.1 documentation
seaborn.histplot — seaborn 0.11.1 documentation

```python
#分布の確認 seabornのsns.displot()やsns.histplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

#分析の下準備
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# print(df.loc[:,'SalePrice'])
sns.displot(df.loc[:,'SalePrice'])
```

```python
#分布の確認 seabornのsns.displot()やsns.histplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

#分析の下準備
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# print(df.loc[:,'SalePrice'])
sns.histplot(df.loc[:,'SalePrice'])
```

```python
#「尖度」と「歪度」も算出
import pandas as pd


var = df.loc[:,'SalePrice'].var() #分散
kurt = df.loc[:,'SalePrice'].kurt() #尖度
skew = df.loc[:,'SalePrice'].skew() #歪度

print("分散：\n", var)
print("尖度：\n", kurt)
print("歪度：\n", skew)
```

```python
#目的変数に対して対数変換を行う。 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。
# 対数変換とは、対数正規分布にしたがう変数の対数をとって正規分布に従う変数を作ることです。
# 対数変換により、少数や、巨大な数値などを整理・表現しやすくすることができます。
# コードで表現したほうがわかりやすいかもしれません。
# data.apply(np.log)
#参考サイト：https://qiita.com/ynakayama/items/2c3782c82317f21596ba
import numpy as np
print(df.loc[:,'SalePrice'].apply(np.log))
sns.displot(df.loc[:,'SalePrice'].apply(np.log))

var = (df.loc[:,'SalePrice'].apply(np.log)).var() #分散
kurt = (df.loc[:,'SalePrice'].apply(np.log)).kurt() #尖度
skew = (df.loc[:,'SalePrice'].apply(np.log)).skew() #歪度

print("分散：\n", var)
print("尖度：\n", kurt)
print("歪度：\n", skew)
```

```python
import numpy as np
# print(df.loc[:,'SalePrice'].apply(np.log))
sns.histplot(df.loc[:,'SalePrice'].apply(np.log))

```

```python
#対数変換の前後での変化を「正規分布」という言葉を含めて説明
#対数変換した分布は、「尖度」と「歪度」の値が小さくなり、正規分布になった。また、平均値が大体真ん中に来るようになった。
```

特徴量の確認をしていきます。それぞれに対してマークダウンで説明や考察を書いてください。


相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。
目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。
選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。

《補足》


重回帰分析などで予測する場合、相関が高すぎる特徴量の組み合わせがあると予測に悪影響が出ることがあります。このような場合の簡単な対処方法として、そういった特徴量の組みを探し出して、１つに絞るということがあります。

```python
#相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

df_corr = df_corr = df.corr()
print(df_corr)

import matplotlib.pyplot as plt
# import japanize_matplotlib
import seaborn as sns
 
sns.heatmap(df_corr)
```

```python
#目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。
# print(df_corr.loc[:,'SalePrice']）
df_corr.sort_values(by='SalePrice', ascending=False).head(11)
sns.heatmap(df_corr.sort_values(by='SalePrice', ascending=False).head(11))
```

```python
#選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。
'''全体的な家の素材と仕上がりの品質
家のリビングの広さ
ガレージサイズ
地下室と１階の広さ
グレード以上のフルバスルーム
グレード以上の部屋の合計
建築された日
改造日
を示している。'''
```

```python
#選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。
print(df_corr.sort_values(by='SalePrice', ascending=False).head(11))
```

```python

df_corr2 = df_corr.sort_values(by='SalePrice', ascending=False).head(11)
print(df_corr2.head(3)) #選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。

df_corr2[df_corr2 > 0.8].stack()
print(df_corr2[df_corr2 > 0.8].stack())
```
