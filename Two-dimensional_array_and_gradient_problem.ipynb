【問題3】勾配を求める
各x座標における勾配を求めます。ここで勾配はyの変化量をxの変化量で割ることで計算します。




```python
#テキストにある通り、xの変化量Δx＝xのスライスの差
#xのスライスの差　= x[1]-x[-1]
Δx = x[1:] - x[:-1]
print(Δx)
#テキストにある通り、yの変化量Δy＝yのスライスの差
#yのスライスの差　= y[1]-y[-1]
Δy=y[1:]-y[:-1]
print(Δy)
#勾配z＝Δy/Δx
z=Δy/Δx
print(z)
```
