【問題1】100日目の米粒の数
100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

```python
import matplotlib.pyplot as plt
sum = 0
sum2 = 0
daylist = []
ricelist = []
ricetotal = []
for i in range(100):
  daylist.append(i)
  sum = 2**(i)
  ricelist.append(sum)
  sum2 = sum2 + sum
  ricetotal.append(sum2)
# return daylist
# print(daylist)
# print(ricelist)
# print(ricetotal)
plt.title("Number and cumulative total of rice received")#もらう米の数と累計
plt.xlabel("Number of days [days]")#日数　[日]
plt.ylabel("Number of grains [grains]")#粒数[粒]
plt.plot(daylist, ricelist, marker="o", color = "red", linestyle = "--")
plt.plot(daylist, ricetotal, marker="v", color = "blue", linestyle = ":");
```

【問題3】貰える米で何人が何日生活できるか
この米があれば何人が何日生活できるか計算する関数を作成してください。コースを受講する期間に貰える米粒で、このクラスにいる人数であれば、何日間生活できるかを出力してみましょう。


関数の最低限の入力は 米粒の数(int) と 人数(int) 、出力は 生活できる日数(int) です。関数名やコメントなどは分かりやすく考えて作成してください。


ここでは1日に必要な米粒の数は提示しません。自分なりに調べて、計算式に落とし込んでください。どう調べ、何を根拠にしたかもコメントを残しておいてください。

```python
'''#https://suiteidan.com/suitei-3/#toc2
のサイトに日本人が1日に食べる米粒は、958,900,000,000 ［粒］とあります。
日本人の人口を120,000,000［人］として、一人あたりが1日に食べる米粒は、約7991[粒/人]であるとする
コース受講する期間：4ヶ月（120日）、クラス人数１１人とする。
4ヶ月でもらえる米粒は、前問から1329227995784915872903807060280344575[粒]
関数名：サバイバル期間　Survival period (略してSP)、
入力：米粒の数(int)Number of rice grains(略してNORG)、
人数Number of People(int)(略してNOP)
出力：生活できる日数Number of days you can live(int)（略してNODYCL）とする'''
def SP(NORG,NOP) :
  NODYCL = NORG/(NOP * 7991)
  return NODYCL

NODYCL = SP(1329227995784915872903807060280344575,11)
print(NODYCL)


```

11人が1.5×１０の３１乗日生きることができる。
