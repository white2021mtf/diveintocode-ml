【問題】栗まんじゅうが太陽系を覆う日

```python
'''太陽系半径(Solar system radius＝SSRと置く)1.575光年(15兆km)=15兆×10^3m
太陽系を球として体積は、4/3πr＾３で求まる。
栗饅頭の体積（Volume of kuri manju=VOKMと置く）は、
10×10^-4m^3とする。X時間で太陽系が栗饅頭で一杯
になるとすると。'''
import math

def KuriManju(SSR,VOKM):
  X = (-math.log(VOKM,10) + math.log(((4/3) * math.pi * (SSR) ** 3),10))/(12*math.log(2,10))
  return X

X = KuriManju(15 * (10 ** 12) * (10 ** 3),10 ** (-4))
print(X)

```

よって、約14.7時間で太陽系は、栗まんじゅうで一杯になる。
